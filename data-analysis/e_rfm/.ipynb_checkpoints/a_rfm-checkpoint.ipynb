```python
import pandas as pd

cp_df = pd.read_csv('./datasets/customer_personality_analysis.csv', sep='\t')
cp_df
```

```python
cp_df.info()
```

```python
cp_df.duplicated().sum()
```

```python
cp_df.isna().sum()
```

```python
cp_df[['Income']].describe().T
```

```python
cp_df['Income'].fillna(cp_df.Income.median(), inplace=True)
cp_df.isna().sum()
```

```python
cp_df.Year_Birth.min()
```

```python
cp_df = cp_df[cp_df.Year_Birth >= 1925]
cp_df
```

```python
cp_df[cp_df.Income <= 0]
```

```python
import numpy as np

Q1 = cp_df.describe().T.loc['Income', '25%']
Q3 = cp_df.describe().T.loc['Income', '75%']

iqr = Q3 - Q1

lower_bound = Q1 - iqr * 1.5
upper_bound = Q3 + iqr * 1.5

if lower_bound < 0:
    lower_bound = 0

print(f'Ï†ïÏÉÅÏπò Î≤îÏúÑ: {lower_bound} ~ {upper_bound}')
```

```python
cp_df = cp_df[cp_df.Income <= upper_bound]
cp_df
```

- Z_CostContact Î∞è Z_Revenue Ïó¥Ïóê ÎåÄÌïú ÏÑ§Î™ÖÏù¥ ÏóÜÏúºÎ©∞ IDÎèÑ ÏÇ¨Ïö©ÎêòÏßÄ ÏïäÏùå
- Complain, AcceptedCmp3, AcceptedCmp2, AcceptedCmp1, AcceptedCmp4,  
  AcceptedCmp5 Î∞è Response Ïó¥Ïùò ÏùëÎãµÎ•†Ïù¥ Îß§Ïö∞ ÎÇÆÏùå

```python
cp_df[['Complain', 'AcceptedCmp3', 'AcceptedCmp2', 'AcceptedCmp1', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']].value_counts()
```

```python
drop_columns = ['Complain', 'AcceptedCmp3', 'AcceptedCmp2', 'AcceptedCmp1', 'AcceptedCmp4', 'AcceptedCmp5', 'Response', 'Z_CostContact', 'Z_Revenue']
cp_df = cp_df.drop(labels=drop_columns, axis=1)
cp_df
```

- Year_BirthÎäî ÌòÑÏû¨ Ïó∞ÎèÑÏóêÏÑú Ï∞®Í∞êÌïòÏó¨ AgeÎ°ú Î≥ÄÌôò

```python
cp_df['Age'] = 2025 - cp_df['Year_Birth']
cp_df
```

```python
cp_df = cp_df.drop(labels=['Year_Birth'], axis=1)
```

- Dt_CustomerÎäî Í≥†Í∞ùÏúºÎ°ú Îì±Î°ùÎêú ÎÇ†ÏßúÏù¥Î©∞, ÌòÑÏû¨ Ïó∞ÎèÑÏóêÏÑú Ï∞®Í∞êÌïòÏó¨ Year_CustomerÎ°ú Î≥ÄÌôòÌïòÍ≥†,   
  Ïó∞Ï∞® Í≥ÑÏÇ∞ÏùÑ ÏúÑÌï¥ datetime typeÏúºÎ°ú Î≥ÄÌôò(ÌòÑÏû¨ Ïó∞ÎèÑÎ•º 2025ÎÖÑÏúºÎ°ú Í∞ÄÏ†ïÌïúÎã§).

```python
cp_df['Dt_Customer'] = pd.to_datetime(cp_df.Dt_Customer, dayfirst=True)
cp_df.sort_values(['Dt_Customer'], ascending=False)
```

```python
cp_df['Year_Customer'] = 2025 - cp_df['Dt_Customer'].dt.year
cp_df
```

```python
cp_df = cp_df.drop(labels=['Dt_Customer'], axis=1)
```

- KidhomeÍ≥º TeenhomeÏùÑ Í≤∞Ìï©ÌïòÏó¨ ChildhomeÏúºÎ°ú Î≥ÄÌôò

```python
cp_df['Childhome'] = cp_df['Kidhome'] + cp_df['Teenhome']
cp_df
```

```python
cp_df = cp_df.drop(labels=['Kidhome', 'Teenhome'], axis=1)
```

- Education Ïª¨ÎüºÏùò Îç∞Ïù¥ÌÑ∞Î•º ÌïôÏúÑ Í∏∞Ï§ÄÏúºÎ°ú Î™ÖÏπ≠ Î≥ÄÍ≤Ω

```python
cp_df['Education'].value_counts()
```

```python
cp_df['Education'] = cp_df['Education'].replace(
    ['2n Cycle', 'Basic', 'Graduation'], ['Master', 'No-degree', 'Bachelor'])

cp_df['Education'].value_counts()
```

- Marital_Status Ïª¨ÎüºÏùò Îç∞Ïù¥ÌÑ∞ Ï§ë Í≤∞Ìòº Ïó¨Î∂ÄÏôÄ Í¥ÄÎ†® ÏóÜÎäî Í≤É ÏÇ≠Ï†ú

```python
cp_df.Marital_Status.value_counts()
```

```python
cp_df = cp_df[~ cp_df['Marital_Status'].isin(['Alone', 'Absurd', 'YOLO'])]
cp_df
```

```python
cp_df.reset_index(drop=True, inplace=True)
cp_df
```

- Í∞Å Íµ¨Îß§ ÌöüÏàò Í¥ÄÎ†® Îç∞Ïù¥ÌÑ∞Î•º Ìï©ÌïòÏó¨ FrequencyÎ°ú Î≥ÄÌôò

```python
frequency_columns = ['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases'] 
cp_df['Frequency'] = cp_df[frequency_columns].sum(axis=1)
cp_df
```

- Í∞Å ÏßÄÏ∂ú Í∏àÏï° Í¥ÄÎ†® Îç∞Ïù¥ÌÑ∞Î•º Ìï©ÌïòÏó¨ MonetaryÎ°ú Î≥ÄÌôò

```python
monetary_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
cp_df['Monetary'] = cp_df[monetary_columns].sum(axis=1)
cp_df
```

```python
cp_rfm_df = cp_df[['Recency', 'Frequency', 'Monetary']]
cp_rfm_df
```

#### Ï†ïÍ∑úÌôî (Nomalization)   
Í∞íÏùò Î≤îÏúÑÎ•º 0~1 ÏÇ¨Ïù¥Î°ú Î≥ÄÌôòÏãúÏºú Î™®Îì† Ïª¨ÎüºÏùò Îç∞Ïù¥ÌÑ∞Í∞Ä ÌèâÎì±Ìï¥ÏßÑÎã§.
ÏÑúÎ°ú Îã§Î•∏ Îã®ÏúÑÏùò Í∞íÏùÄ ÎπÑÍµê ÎåÄÏÉÅÏù¥ Îê† Ïàò ÏóÜÎã§. ÏòàÎ•º Îì§Ïñ¥, 80kgÍ≥º 180cmÎäî ÎπÑÍµêÌï† Ïàò ÏóÜÍ∏∞ ÎïåÎ¨∏Ïóê Ï†ïÍ∑úÌôîÎ•º ÏÇ¨Ïö©Ìï¥ÏÑú ÎπÑÍµêÌïúÎã§.

```python
# !pip install scikit-learn
```

```python
from sklearn.preprocessing import MinMaxScaler

normalization = MinMaxScaler()
rfm_normalization = normalization.fit_transform(cp_rfm_df)
rfm_normalization
```

```python
cp_rfm_norm_df = pd.DataFrame(rfm_normalization, columns=cp_rfm_df.columns)
cp_rfm_norm_df
```

```python
cp_rfm_norm_df['Recency'] = 1 - cp_rfm_norm_df['Recency']
cp_rfm_norm_df
```

```python
cp_df['Total_Score'] = cp_rfm_norm_df['Recency'] + cp_rfm_norm_df['Frequency'] + cp_rfm_norm_df['Monetary']
cp_df
```

```python
cp_result_df = cp_df[['ID', 'Education', 'Marital_Status', 'Income', 'Age', 'Childhome', 'Total_Score']]
cp_result_df
```

```python
import numpy as np
g4, g3, g2, g1 = np.percentile(cp_df['Total_Score'], [20, 40, 70, 90])
print(g4, g3, g2, g1)
```

```python
def get_grade(x):
    if x <= g4:
        return 5
    if x <= g3:
        return 4
    if x <= g2:
        return 3
    if x <= g1:
        return 2
    return 1
```

```python
cp_df['Grade'] = cp_df['Total_Score'].apply(get_grade)
cp_df
```

```python
cp_df['Badge'] = cp_df['Grade'].replace([1, 2, 3, 4, 5], ['VIP', 'Diamond', 'Gold', 'Silver', 'Bronze'])
cp_df
```

```python
cp_df['Badge'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

order_columns = ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP']

sns.countplot(x='Badge', data=cp_df, palette='muted', order=order_columns)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

sns.boxplot(ax=axes[0], data=cp_df, x='Badge', y='NumWebVisitsMonth', palette='muted', order=order_columns)
sns.boxplot(ax=axes[1], data=cp_df, x='Badge', y='Age', palette='muted', order=order_columns)
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 5))

sns.countplot(ax=axes[0], data=cp_df, x='Badge', hue='Education', palette='muted', order=order_columns)
sns.countplot(ax=axes[1], data=cp_df, x='Badge', hue='Marital_Status', palette='muted', order=order_columns)
sns.countplot(ax=axes[2], data=cp_df, x='Badge', hue='Childhome', palette='muted', order=order_columns)
plt.show()
```
