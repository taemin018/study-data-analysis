### RFM 분석 실습
- Recency: 얼마나 최근에 구매했는가
- Frequency: 얼마나 자주 구매했는가
- Monetary: 얼마나 많은 금액을 지출했는가

### 고객 분석
- InvoiceNo: 거래 번호
- StockCode: 상품 코드
- Description: 상품명 
- InvoiceDate: 거래 날짜
- UnitPrice: 상품 단가
- CustomerID: 고객 번호
- Country: 국가명

```python
import pandas as pd

ci_df = pd.read_csv('./datasets/customer_invoice_data.csv')
ci_df
```

```python
ci_df.isna().sum()
```

```python
ci_df = ci_df.drop(index=ci_df[ci_df.CustomerID.isna()].index, axis=0)
ci_df
```

```python
ci_df = ci_df[ci_df.UnitPrice > 0]
ci_df
```

```python
rfm_f = ci_df.CustomerID.value_counts().to_frame().reset_index()
rfm_f
```

```python
rfm_f.columns = ['CustomerID', 'Frequency']
rfm_f
```

```python
ci_df['Monetary'] = ci_df.Quantity * ci_df.UnitPrice
ci_df
```

```python
ci_df.InvoiceDate = pd.to_datetime(ci_df.InvoiceDate)
ci_df
```

```python
ci_df.sort_values(by='InvoiceDate', ascending=False)
```

```python
ci_df.loc[:, 'Recency'] = pd.to_datetime(ci_df['InvoiceDate'])\
.apply(lambda x: (pd.to_datetime('2011-12-10') - x).days)

ci_df
```

```python
ci_df = ci_df.drop(['InvoiceDate'], axis=1)
```

```python
rfm = ci_df.groupby('CustomerID').agg(Recency=('Recency', 'min'), Frequency=('CustomerID', 'count'), Monetary=('CustomerID', 'sum'))
rfm = rfm.reset_index()
rfm
```

```python
from sklearn.preprocessing import MinMaxScaler

rfm_scale = MinMaxScaler()
rfm_scale = rfm_scale.fit_transform(rfm[['Recency', 'Frequency', 'Monetary']])
```

```python
rfm[['Recency', 'Frequency', 'Monetary']] = pd.DataFrame(rfm_scale, columns=['Recency', 'Frequency', 'Monetary'])
rfm
```

```python
rfm['Recency'] = 1 - rfm['Recency']
rfm
```

```python
rfm.loc[:, 'TotalScore'] = rfm['Recency'] + rfm['Frequency'] + rfm['Monetary']
rfm
```

```python
ci_rfm_df = ci_df.merge(rfm[['CustomerID', 'TotalScore']], on='CustomerID')
ci_rfm_df
```

```python
import numpy as np

l1, l2, l3, l4 = np.percentile(ci_rfm_df['TotalScore'], [20, 40, 70, 90])
print(l1, l2, l3, l4)
```

```python
def get_level(x):
    if x <= l1:
        return 5
    if x <= l2:
        return 4
    if x <= l3:
        return 3
    if x <= l4:
        return 2
    return 1
```

```python
ci_rfm_df['Level'] = ci_rfm_df['TotalScore'].apply(get_level)
ci_rfm_df
```

```python
ci_rfm_df['Level'] = ci_rfm_df['Level'].replace([5, 4, 3, 2, 1], 
                                                ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

order = ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP']

sns.countplot(x = 'Level', data=ci_rfm_df, palette='muted', order=order)
plt.title('Level Count')
plt.show()
```
