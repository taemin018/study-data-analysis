#### DataFrame()
- dict를 DataFrame으로 변환하고자 할 때 DataFrame 생성자에 전달한다.
- 컬럼명을 추가하거나 인덱스명을 변경하는 등 다양하게 설정할 수 있다.

```python
import pandas as pd

film = {
    'title': ['명량', '극한 직업', '범죄 도시3', '국제 시장'],
    'audience': [17_615_919, 16_266_480, 10_682_674, 14_265_222],
    'country': ['한국', '한국', '한국', '한국']
}

film_df = pd.DataFrame(film)
display(film_df)
```

```python
# feature 추가
film_df['income'] = [135_758_658_810, 139_657_105_516, 104_686_489_632, 110_951_970_230]
display(film_df)
```

```python
# index 수정
film_df.index = ['one', 'two', 'three', 'four']
display(film_df)
```

```python
# 인덱스를 0부터 순서대로 다시 세팅
# drop: 기존 인덱스 삭제 여부
# inplace: 원본 Dataframe에 적용 여부
film_df.reset_index(drop=False, inplace=True)
film_df
```

```python
# feature 삭제
film_df.drop(labels=['index'], axis=1, inplace=True)

# 행 삭제
# film_df.drop(index=[0], axis=0)

film_df
```

```python
# feature 이름 수정
# {'기존이름': '새로운이름'}
film_df.rename(columns={'title': 'name'})
```

#### describe()
- 숫자형 데이터의 개수, 평균, 표준편차, 최소값, 사분위 분포도(중앙값: 50%), 최대값을 제공한다.
- 25번째 백분위수와 75번째 백분위수를 기준으로 정상치의 범위를 설정할 수 있다.

```python
happiness_df = pd.read_csv('./datasets/happiness_report_2022.csv')
happiness_df.describe().T
```

```python
import numpy as np
from scipy.stats import iqr

happiness_Q1 = np.percentile(happiness_df.score, 25)
happiness_Q3 = np.percentile(happiness_df.score, 75)
# print(happiness_Q1, happiness_Q3)

# iqr_value = happiness_Q3 - happiness_Q1
# print(iqr_value)

iqr_value = iqr(happiness_df.score)
print(iqr_value)

lower_bound = happiness_Q1 - 1.5 * iqr_value
upper_bound = happiness_Q3 + 1.5 * iqr_value
print(f'정상치 범위: {lower_bound} ~ {upper_bound}')
```

```python
condition1 = happiness_df.score >= lower_bound
condition2 = happiness_df.score <= upper_bound
condition = condition1 & condition2

happiness_df = happiness_df[condition]
happiness_df
```

```python
happiness_df.isna().sum()
```

```python
happiness_df.drop(index=happiness_df[happiness_df.income.isna()].index, axis=0, inplace=True)
happiness_df
```

```python
happiness_df.reset_index(drop=True, inplace=True)
happiness_df
```

```python
happiness_df.duplicated().sum()
# happiness_df.drop_duplicates()
```

```python
hp_mean_df = happiness_df.groupby('income').mean('score').reset_index()
hp_mean_df
```

```python
hp_min_df = happiness_df.groupby('income').min('score').reset_index()
hp_min_df
```

```python
hp_max_df = happiness_df.groupby('income').max('score').reset_index()
hp_max_df
```

```python
hp_std_df = happiness_df.groupby('income')['score'].std().reset_index()
hp_std_df
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(12, 9))

hp_mean_df.plot(title='mean', x='income', y='score', fontsize='7', kind='line', grid=True, ax=axes[0, 0])
hp_min_df.plot(title='min', x='income', y='score', fontsize='7', kind='line', grid=True, ax=axes[0, 1])
hp_max_df.plot(title='max', x='income', y='score', fontsize='7', kind='line', grid=True, ax=axes[1, 0])
hp_std_df.plot(title='std', x='income', y='score', fontsize='7', kind='line', grid=True, ax=axes[1, 1])
```

- "Upper middle income" (중상위 소득) 그룹의 표준편차는 0.634569로, 중하위 소득 그룹보다는 약간 작지만
