## 행복지수와 GDP 관계 그래프 해석 
gdp_per_capita(1인당 국내총생산(GDP)을 측정)에 따른 행복지수 그래프 
x축은 gdp_per_capita(1인당 국내총생산(GDP), y축은 행복지수 점수

- GDP가 높을 수록 행복지수 점수가 높다.
- GDP가 낮은 국가들(왼쪽 구간, 0 ~ 0.8)은 대체로 행복 점수가 3~5점 수준으로 낮다.
- GDP가 중간 정도인 국가들(1.0 ~ 1.5)은 행복 점수가 5~6점 사이에서 분포한다.
- GDP가 높은 국가들(1.8 이상)은 대부분 행복 점수가 7점 이상이다.

그래프가 우상향하는 것을 보면 경제 수준이 높은 나라일수록 국민들의 행복감이 높게 나타난다.

```python
line = sns.lineplot(data=happiness_df, x='region', y='perceptions_of_corruption')
line.set_xticklabels(line.get_xticklabels(), rotation=90)
plt.show()
```

#### 그래프 보고서 
지역별 perceptions_of_corruption(국가 내 부패에 대한 인식) 수치
- y축은 국가 내 부패에 대한 인식 수치, x축은 지역

- 서유럽(Western Europe) 과 북미 및 오세아니아(North America and ANZ) 지역은 평균적으로 가장 높은 부패 인식도(약 0.3 이상)를 보였다.
이는 해당 지역의 정부 투명성과 제도적 신뢰도가 높으며, 부패가 적다고 인식되는 경향을 의미한다.

- 중동 및 북아프리카(Middle East and North Africa),
사하라 이남 아프리카(Sub-Saharan Africa), 동유럽(Central and Eastern Europe) 지역은 0.1 이하의 낮은 수치를 보여, 부패가 만연하다고 느끼는 정도가 높고,
정부 및 제도에 대한 신뢰가 상대적으로 낮은 편으로 해석된다.

- 동남아시아(Southeast Asia) 와 동아시아(East Asia) 지역은 중간 수준(약 0.15~0.25)의 값을 보여,
국가별로 청렴도 격차가 크며 사회적 신뢰 수준이 다양하게 분포하는 것으로 보인다.

결론은 부패에 대한 인식수치가 높을 수록 치안이 좋다는 것을 알 수 있다.
